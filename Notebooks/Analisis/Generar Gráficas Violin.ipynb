```python
import pandas as pd
import numpy as np
import os
import re

import matplotlib.pyplot as plt
```

## Cargar Datos

```python
daily_data_sheet_path = os.path.join('..','..','dataset','Series','Sabanas','Original','Sabana_Datos_Diaria.xlsx')
daily_data_sheet = pd.read_excel(daily_data_sheet_path)
daily_data_sheet = daily_data_sheet.set_index('Fecha')
```

```python
hourly_data_sheet_path = os.path.join('..','..','dataset','Series','Sabanas','Original','Sabana_Datos_Horaria.xlsx')
hourly_data_sheet = pd.read_excel(hourly_data_sheet_path)
hourly_data_sheet = hourly_data_sheet.set_index('Fecha')
```

## Generar y Guardar Gráficas

```python
def guardar_graficas(x,temporalidad):
    for variable in x.columns:
        fig, axis = plt.subplots(figsize=(8,5))
        
        img_path = os.path.join('Gráficos de Violin',temporalidad)
        if not os.path.exists(img_path):
            os.makedirs(img_path) 
        
        img_path = os.path.join(img_path,'{}.jpg'.format(re.sub(r'[\W]', ' ', variable)))
        
        plt.violinplot(x[variable])        
        plt.title(variable)
        plt.savefig(img_path)
```

```python
guardar_graficas(x=daily_data_sheet,temporalidad='Diario')
```

```python
guardar_graficas(x=hourly_data_sheet,temporalidad='Horario')
```
